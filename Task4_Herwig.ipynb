```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Top Absolute Correlations")
print(get_top_abs_correlations(correlation, 50))
```

```python
def get_bottom_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=True)
    return au_corr[0:n]

print("Bottom Absolute Correlations")
print(get_bottom_abs_correlations(correlation, 50))
```

```python
def get_all_abs_correlations(df):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=True)
    return au_corr
ordered_corr = get_all_abs_correlations(correlation)

print(ordered_corr[:60])
```

```python
def get_all_abs_correlations_desc(df):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr
ordered_corr_desc = get_all_abs_correlations_desc(correlation)

print(ordered_corr_desc[:60])
```

<h3>Define the KNN Classifier</h3>

```python
def test_model(model, X_test, y_test):
    model.fit(X_test, y_test)
    # Predict data
    prediction = model.predict(X_test)
    #check accuracy of our model on the test data
    score = model.score(X_test, y_test)
    
    print('Score: {}'.format(score))
```

```python
def KNNClassifier(X, y, neighbors = [], splits = [], random_state = 2, shuffle = True):
    best_model = None
    previous_score = None
    training_score = None
    best_neighbor = None
    best_split = None
    predictions = None
    test_data = None
    for split in splits:
        kf = StratifiedKFold(n_splits = split, shuffle = shuffle, random_state = random_state)
        for neighbor in neighbors:
            for train_index, test_index in kf.split(X, y):
                X_train, X_test = X[train_index], X[test_index]
                y_train, y_test = y[train_index], y[test_index]

                # Create KNN classifier
                knn = KNeighborsClassifier(n_neighbors = neighbor)

                # Fit the classifier to the data
                model = knn.fit(X_train,y_train)

                # Predict data
                prediction = model.predict(X_test)
                
                # check accuracy of our model on the test data
                score = model.score(X_test, y_test)
                # score on training data
                training_score = model.score(X_train, y_train)

                # save the best model
                if best_model is None:
                    best_model = model
                    previous_score = score
                    best_neighbor = neighbor
                    best_split = split
                    training_score = training_score
                    predictions = prediction
                    test_data = y_test
                elif score > previous_score:
                    best_model = model
                    previous_score = score
                    best_neighbor = neighbor
                    best_split = split
                    training_score = training_score
                    predictions = prediction
                    test_data = y_test
                    
        print('- Split {} -'.format(split))
        print('Best accuracy on Training set {}'.format(training_score))
        print('Best accuracy on Test set {}'.format(previous_score))
        print('Best neighbor {}'.format(best_neighbor))
        #print('Best split {}'.format(best_split))
        print('Accuracy on test set (data not used in any split)')
        test_model(best_model, x_test_gen, y_test_gen)
        print('=========================================')
        
        # Set up the matplotlib figure
        f, ax = plt.subplots(figsize=(30, 20)) ## here increased figsize

        ## HERE - max tick font size
        ax.tick_params(axis='both', labelsize=30)

        mat = confusion_matrix(y_test, prediction)
        sns_plot = sns.heatmap(mat.T, annot=True, linewidths=.5, square=False, fmt='d', cbar=False, annot_kws={"fontsize":40})
        f.set_tight_layout(True)
        plt.title('Split: {} | Training accuracy: {} | Test accuracy: {} | Best neighbor: {}'.format(split, training_score, previous_score, best_neighbor), fontsize = 40) # title with fontsize 20
        plt.xlabel('Predicted Label', fontsize = 50)
        plt.ylabel('True Label', fontsize = 50)
        plt.show()

                
    print('Overall best model score {}'.format(previous_score))
    print('Overall best neighbor {}'.format(best_neighbor))
    print('Overall best split {}'.format(best_split))
    
    return (best_model, previous_score, best_neighbor, best_split, mat)
```

```python
def KNNClassifier_Gain(X, y, neighbors = [], splits = [], random_state = 2, shuffle = True, plot = False):
    best_model = None
    previous_score = None
    training_score = None
    best_neighbor = None
    best_split = None
    predictions = None
    test_data = None
    best_revenue = None
    
    for split in splits:
        kf = StratifiedKFold(n_splits = split)
        for neighbor in neighbors:
            for train_index, test_index in kf.split(X, y):
                X_train, X_test = X[train_index], X[test_index]
                y_train, y_test = y[train_index], y[test_index]

                # Create KNN classifier
                knn = KNeighborsClassifier(n_neighbors = neighbor)

                # Fit the classifier to the data
                model = knn.fit(X_train,y_train)

                # Predict data
                prediction = model.predict(X_test)
                
                # check accuracy of our model on the test data
                score = model.score(X_test, y_test)
                # score on training data
                training_score = model.score(X_train, y_train)
                
                gain_matrix = np.array([
                    [5, -5, -5, 2],
                    [-5, 10, 2, -5],
                    [-5, 2, 10, -5],
                    [2, -5, -2, 5]])
                
                confusions = confusion_matrix(y_test, prediction)
                
                revenue = (confusions * gain_matrix).sum()
                

                # save the best model with highest gain
                if best_model is None:
                    best_model = model
                    previous_score = score
                    best_neighbor = neighbor
                    best_split = split
                    training_score = training_score
                    predictions = prediction
                    test_data = y_test
                    best_revenue = revenue
                elif revenue > best_revenue:
                    best_model = model
                    previous_score = score
                    best_neighbor = neighbor
                    best_split = split
                    training_score = training_score
                    predictions = prediction
                    test_data = y_test
                    best_revenue = revenue
                    
        print('- Split {} -'.format(split))
        print('Best accuracy on Training set {}'.format(training_score))
        print('Best accuracy on Test set {}'.format(previous_score))
        print('Best neighbor {}'.format(best_neighbor))
        print('Best revenue {}'.format(best_revenue))
        #print('Best split {}'.format(best_split))
        print('Accuracy on test set (data not used in any split)')
        test_model(best_model, x_test_gen, y_test_gen)
        print('=========================================')
        
        if plot:
            # Set up the matplotlib figure
            f, ax = plt.subplots(figsize=(30, 20)) ## here increased figsize

            ## HERE - max tick font size
            ax.tick_params(axis='both', labelsize=30)

            mat = confusion_matrix(y_test, prediction)
            sns_plot = sns.heatmap(mat.T, annot=True, linewidths=.5, square=False, fmt='d', cbar=False, annot_kws={"fontsize":40})
            f.set_tight_layout(True)
            plt.title('Split: {} | Training accuracy: {} | Test accuracy: {} | Best neighbor: {}'.format(split, training_score, previous_score, best_neighbor), fontsize = 40) # title with fontsize 20
            plt.xlabel('Predicted Label', fontsize = 50)
            plt.ylabel('True Label', fontsize = 50)
            plt.show()

                
    print('Overall best model score {}'.format(previous_score))
    print('Overall best neighbor {}'.format(best_neighbor))
    print('Overall best split {}'.format(best_split))
    
    return (best_model, previous_score, best_neighbor, best_split)
```
